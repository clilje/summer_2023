```python
import numpy as np
import matplotlib.pyplot as plt
import math
import csv
import pandas as pd
import os
```

```python
df = pd.read_csv('ASCII/bc2003_lr_BaSeL_m22_chab_ssp.130','/s', header=None, skiprows=6)
df = df[0].str.split(expand=True)

wavelength = df[0].astype('float64').to_numpy()
flux = df[1].astype('float64').to_numpy()
plt.plot(wavelength,flux, label='0.5')
flux = df[2].astype('float64').to_numpy()
plt.plot(wavelength,flux,label='1')
flux = df[3].astype('float64').to_numpy()
plt.plot(wavelength,flux,label='2')
flux = df[4].astype('float64').to_numpy()
plt.plot(wavelength,flux,label='5')
flux = df[5].astype('float64').to_numpy()
plt.plot(wavelength,flux,label='10')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```

```python
#galaxy at age 0.153 (corresponds to z = 8)
df = pd.read_csv('ASCII/bc2003_lr_BaSeL_m22_chab_ssp.120','/s', header=None, skiprows=6)
df = df[0].str.split(expand=True)
#distance modulus at z = 8
mu = 49.53
wavelength = df[0].astype('float64').to_numpy()
flux = df[1].astype('float64').to_numpy()
print(wavelength,flux)

flux = 3.826e+33*flux
flux = (flux)/((4*np.pi*(10**((mu/5)+1))**2))
flux = flux*(0.00000000000000000032408**2)
print(flux)
flux = flux*(3*10e+18)/(wavelength**2)
print(flux)
wavelength = 0.0001*wavelength
plt.plot(wavelength,flux*10e+20)
plt.xscale('log')
plt.yscale('log')
plt.xlim(1.1,5)
#plt.ylim(10e-3,10e+1)
plt.xlabel(r'$\mu$ m')
plt.ylabel(r'$f_{\lambda}$ ( erg/s/A/cm^2)')
plt.show()
```

```python
#galaxy at age 0.153 (corresponds to z = 8)
df = pd.read_csv('ASCII/m22','/s', header=None, skiprows=6)
df = df[0].str.split(expand=True)
print(df)
names = ['0.001','0.01','0.1','0.4','1','4','13']
#distance modulus at z = 8
mu = 49.53
wavelength = df[0].astype('float64').to_numpy()
wavelength = 0.0001*wavelength
i = 1
while i < 8:

    flux = df[i].astype('float64').to_numpy()

    flux = 3.826e+33*flux
    flux = (flux)/((4*np.pi*(10**((mu/5)+1))**2))
    flux = flux*(0.00000000000000000032408**2)
    #print(flux)
    flux = flux*(3*10e+18)
    #print(flux)
    plt.plot(wavelength,flux*10e+20,label=names[i-1])
    i+=1
plt.xscale('log')
plt.yscale('log')
plt.ylim(10e-6,40)
#plt.ylim(10e-3,10e+1)
plt.legend()
plt.xlabel(r'Wavelength in $\mu$ m')
plt.ylabel(r'$f_{\lambda}$ ($10^{-20}$ erg/s/A/cm^2)')
plt.savefig('Figures/sed_different_age')
plt.show()
0.001,0.01,0.1,0.4,1,4,13
```

```python
cols = 4 
rows = 2
i = 0
j = 0
fig, axes = plt.subplots(rows, cols, figsize=(16, 12))
files = ['ASCII/m22','ASCII/m32','ASCII/m42','ASCII/m52','ASCII/m62','ASCII/m72','ASCII/m82']
names = ['0.001','0.01','0.1','0.4','1','4','13']
x = 0
while x < len(files):
    #galaxy at age 0.153 (corresponds to z = 8)
    print(files[x])
    df = pd.read_csv(files[x],'/s', header=None, skiprows=6)
    df = df[0].str.split(expand=True)
    print(df)
    #distance modulus at z = 8
    mu = 49.53
    wavelength = df[0].astype('float64').to_numpy()
    wavelength = 0.0001*wavelength
    f = 1
    while f < 8:
        flux = df[f].astype('float64').to_numpy()
        flux = 3.826e+33*flux
        flux = (flux)/((4*np.pi*(10**((mu/5)+1))**2))
        flux = flux*(0.00000000000000000032408**2)
        #print(flux)
        flux = flux*(3*10e+18)
        #print(flux)
        axes[i][j].plot(wavelength,flux*10e+20,label=names[f-1])
        f+=1
    axes[i][j].set_xscale('log')
    axes[i][j].set_yscale('log')
    axes[i][j].set_title(files[x])
    axes[i][j].set_ylim(10e-6,40)
    axes[i][j].set_xlim(0.01,100)
    axes[i][j].legend()
    axes[i][j].set_xlabel(r'Wavelength in $\mu$ m')
    axes[i][j].set_ylabel(r'$f_{\lambda}$ ($10^{-20}$ erg/s/A/cm^2)')
    x += 1
    j += 1
    if x == 4:
        i +=1
        j = j-4
plt.tight_layout()
plt.savefig('Figures/sed_different_metallicity')
plt.show()
0.001,0.01,0.1,0.4,1,4,13
```

These following cells are all concerned with plotting different star formation histories.

They are based on the .4color output files of the BC03 code version 2016.

The different cells use a different star formation history. 
We can clearly see that for custom models there is linear interpolation. 

```python
df = pd.read_csv('SFR-trials/sfrSSP.txt', delimiter= '\s+', skiprows=30,names=['log-age-yr','Mbol','Bmag',
                                                                             'Vmag','Kmag','M*_liv','M_remnants',
                                                                             'M_ret_gas','M_galaxy','SFR/yr',
                                                                             'M*_liv+M_rem','M*_tot/Lb','M*_tot/Lv',
                                                                             'M*_tot/Lk','M*_liv/Lb','M*_liv/Lv<','M*_liv/Lk'])
print(df)
plt.plot(df['log-age-yr'].to_numpy(), df['M*_liv+M_rem'].to_numpy(), label='M_star')
plt.plot(df['log-age-yr'].to_numpy(), df['M_galaxy'].to_numpy(), label='M_galaxy')
plt.plot(df['log-age-yr'].to_numpy(), df['SFR/yr'].to_numpy(), label='SFR/yr')
plt.legend()
plt.xlabel('Log age in years')
plt.savefig('SSPSFR0')
plt.show()
```

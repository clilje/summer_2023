```python
import numpy as np
import matplotlib.pyplot as plt
import math
import csv
import pandas as pd
import os
import matplotlib.gridspec as gridspec
```

```python
df = pd.read_csv('TNG_50_library/magnitudes_galaxies.txt', header=None, 
                 names=['ID','redshift','LTT','tz ','dm','F070','F090','F115','F150','F200',
                        'F277','F356','F444','F0560','F0770','F1000','F1130',
                        'F1280','F1500','F1800','F2100','F2550'],dtype=float)
df['color:115-150'] = df['F115']-df['F150']
df['color:150-200'] = df['F150']-df['F200']

df['color:200-277'] = df['F200']-df['F277']
df['color:277-356'] = df['F277']-df['F356']

df['color:356-444'] = df['F356']-df['F444']
print(df)

df2 = pd.read_csv('TNG_50_library/magnitudes_galaxies_with_BH.txt', header=None, 
                 names=['ID','redshift','LTT','tz ','dm','F070','F090','F115','F150','F200',
                        'F277','F356','F444','F0560','F0770','F1000','F1130',
                        'F1280','F1500','F1800','F2100','F2550'],dtype=float)
df2['color:115-150'] = df2['F115']-df2['F150']
df2['color:150-200'] = df2['F150']-df2['F200']

df2['color:200-277'] = df2['F200']-df2['F277']
df2['color:277-356'] = df2['F277']-df2['F356']

df2['color:356-444'] = df2['F356']-df2['F444']



import matplotlib as mtp
mtp.rcParams['figure.figsize']=[20,20]
fig=plt.figure()
gs1=gridspec.GridSpec(2,2)
gs1.update(left=0.1,right=0.5, bottom=0.2, top=0.6, hspace=0.3, wspace=0.3)
ax=plt.subplot(gs1[0])
ax.tick_params(axis='both',which='both',direction='in',bottom=True,top=True,left=True,right=True)
ax.scatter(df['color:150-200'],df['color:115-150'],marker='+', label='galaxies')
ax.scatter(df2['color:150-200'],df2['color:115-150'],marker='+', label='galaxies + BH')
ax.set_ylim(-0.45,0.45)
ax.set_xlim(-0.45,0.45)
ax.set_xlabel('color:150-200')
ax.set_ylabel('color:115-150')
ax.legend()

ax=plt.subplot(gs1[1])
ax.tick_params(axis='both',which='both',direction='in',bottom=True,top=True,left=True,right=True)
ax.scatter(df['color:277-356'],df['color:200-277'],marker='+', label='galaxies')
ax.scatter(df2['color:277-356'],df2['color:200-277'],marker='+', label='galaxies + BH')

ax.set_ylim(-0.45,0.45)
ax.set_xlim(-0.45,0.45)
ax.set_xlabel('color:277-356')
ax.set_ylabel('color:200-277')
ax.legend()

ax=plt.subplot(gs1[2])
ax.tick_params(axis='both',which='both',direction='in',bottom=True,top=True,left=True,right=True)
ax.scatter(df['color:200-277'],df['color:150-200'],marker='+', label='galaxies')
ax.scatter(df2['color:200-277'],df2['color:150-200'],marker='+', label='galaxies + BH')

ax.set_ylim(-0.45,0.45)
ax.set_xlim(-0.45,0.45)
ax.set_xlabel('color:200-277')
ax.set_ylabel('color:150-200')
ax.legend()

ax=plt.subplot(gs1[3])
ax.tick_params(axis='both',which='both',direction='in',bottom=True,top=True,left=True,right=True)
ax.scatter(df['color:356-444'],df['color:277-356'],marker='+', label='galaxies')
ax.scatter(df2['color:356-444'],df2['color:277-356'],marker='+', label='galaxies + BH')

ax.set_ylim(-0.45,0.45)
ax.set_xlim(-0.45,0.45)
ax.set_xlabel('color:356-444')
ax.set_ylabel('color:277-356')
ax.legend()
```
